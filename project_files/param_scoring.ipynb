# Parameter Scoring
This is the final version of the parameter scoring notebook. It uses functions found in the scoring.py file to automatically assign preliminary scores to parameters. These preliminary scores must be manually reviewed for accuracy on a case-by-case basis with extensive knowledge of the models.

```python
# Import libraries
import numpy as np
import pandas as pd
from scoring import load_df, inclusion
```

```python
# Define models and subsets
models = ['aquacrop', 'dssat', 'stics', 'rodekoning', 'rotomgro', 'genericdescriptive']
subsets = ['', 'raw_data', 'raw_trials', 'raw_data_only', 'composite']
```

```python
# For each model and subset combination, load the data and create the scoring sheet
for model in models:
    for subset in subsets:
        df = load_df(model, subset)
        inclusion(df, model, subset)
        print(f'{model} {subset} done')
```
